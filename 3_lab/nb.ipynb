### Тестування

- тестування - це процес перевірки виконання коду з наперед відомим результатом.  При тестування ми виконуємо певні дії над кодом, або проробляємо певні сценарії виконання коду і перевіряємо результат. Цей результат найчастіше є відомим, і ми співставляємо чи отриманий результат відповідає очікуванному. 
- найпростішим видом тестування є перевірка методом `assert` який являє собою набір тверджень, що мають виконуватись.
- така перевірка є схожою на блок `if` однак займає набагато менше коду 

```python
a = True # boolean
b = 1 # Для хибного твердження поставте - False  

assert a == b, "Твердження є хибне" 
# після твердження ми можемо записати текст помилки або пояснення чому є помилка
```

```python
# така перевірка assert повністю підпадає під конструкцію блоку if
if a == b:
    print("Твердження вірне")
else: 
    raise AssertionError("Твердження хибне")
```

> будемо залишити у комірках лише true твердження, щоб не ломати послідовність виконання комірок
> для того, щоб побачити процес виявлення умов які не підходять під твердження ми будемо залишати коменти, де нам потрібно змінити код

### Приклади Assert


```python
c = 1 # number integer
d = "1" # string 
e = 1.0 # float

# зробимо твердження на правильність типів даних для наших змінних 
assert isinstance(a, bool), "Змінна не відноситься до логічного типу даних"
assert isinstance(c, int), "Змінна не відноситься до типу даних - цілих чисел"
assert isinstance(d, str), "Змінна не відноситься до string"

# можемо перевіряти власне значення на відповідність чомусь 
assert c == e, "Значення не співпадають"
# Твердження нижче будуть видавати помилки, тому що:

# не співпадають типи значення, що перевіряються
# assert type(c) is type(e), f"Тип {type(c)} не відповідає {type(e)}"

# не співпадають значення, число !== символу
# assert c == d, f"Значення не є рівними бо {c} не рівне {d}"

# не співпадають типи значення, що перевіряються
# assert type(c) is type(d), f"Тип {type(c)} не відповідає {type(d)}"
```

```python
# може бути твердження не на чітку рівність, а на попадання в певний діапазон 
assert c >= 1, "Значення є меншим за 1"
assert (0 < c < 4), "Умова не відповідає діапазону від 0 до 4"
assert d in [1, 1.0, "1"], "Значення не відповідає вимогам "
```

```python
# ми можемо писати комплексні перевірки у вигляді ф-кцій та використовувати ці ф-кції як твердження
def check(n):
    return n > 0

assert check(c), "перевірка значення в середині функції не є істиною"

def check_numbers(n:list, t):
    """Функція перевіряє чи в списку містяться дані задоного типу"""
    d = [x for x in n if isinstance(x, t)]
    return len(d)

print(type(None))
text_list = [1, "1", 1.0, 2, "2", 2.5, None] # В цьому списку лише 2 значення, які відповідають вимогам
f = check_numbers(text_list, str)
print(f"Перевіряємо вивід функції {f}")
assert f == 2, "Тестовий масив повинен повернути значення 2"
```

```python
# Таку саму логіку з твердженнями ми можемо виконувати на об'єктах
class Test:
    pass

o = Test()
h = Test()
assert isinstance(o, Test)
assert o.__hash__() == hash(o), "Об'єкти мають однаковий Хеш, тобто вони є ідентичні"
# Два об'єкти одного класу не будуть однаковими, бо кожен об'єкт унікальний
# assert o == h, f"{o} не буде ідентичним з {h}"
```
